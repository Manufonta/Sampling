## Ejercicio 1
Coje un conjunto de datos del tema deportivo que te guste. Realiza un muestreo de los datos generando una muestra aleatoria simple y una muestra sistemática.

```python
import numpy as np
import pandas as pd
import random
from sklearn.model_selection import train_test_split
```

```python
liga = pd.read_csv('laliga_player_stats_english.csv')
liga
```

```python
liga.columns
```

Para realizar el ejercicio será suficiente seleccionar solo unas cuantas columnas:

```python
liga = liga[['Team', 'Position', 'Name', 'Minutes played','Goals scored','Passes']]
liga
```

```python
liga.info()
```

```python
liga.describe().round(2)
```

```python
# Muestra aleatoria simple de 50 jugadores

k = 50
liga_sample = liga.sample(k)
```

```python
liga_sample.shape
```

```python
liga_sample.head()
```

```python
# Muestra aleatoria sistemática de 50 jugadores
# k es el número de la muestra; p es el número aleatorio por el que empieza la selección e i es el intervalo.

k = 55
i = int(len(liga)/k)

p = random.randint(0, int(len(liga)))

# DataFrame de los números aleatorios que se seleccionarán.
numbers = pd.DataFrame( p + i*np.arange(k), columns= ['indice'])
numbers['indice'] = numbers['indice'].apply(lambda x: x - len(liga) if x > len(liga) else x)
numbers = numbers.set_index('indice')

#Se utiliza 'numbers' para seleccionar los jugadores del dataframe 'liga', que son ordenados alfabéticamente.
liga_sys_sample = numbers.merge(liga.sort_values('Name').reset_index(), how='inner', left_index=True, right_index=True)
liga_sys_sample
```

```python

liga_sys_sample.shape
```

## Ejercicio 2

Genera una muestra estratificada y una muestra utilizando SMOTE (Synthetic Minority Oversampling Technique).

https://machinelearningmastery.com/smote-oversampling-for-imbalanced-classification/

```python
# Muestra estratificada por posicion
liga_str_sample = liga.groupby('Position').apply(pd.DataFrame.sample, frac = 0.1)
liga_str_sample
```

```python
liga_str_sample.shape
```

A partir de esta muestra, se utilizará el método SMOTE para equilibrar el número de muestras por posición:

```python
from imblearn.over_sampling import SMOTE
```

```python
Samp = np.array(liga_str_sample[['Minutes played','Passes']])
y = np.array(liga_str_sample['Position'])
```

```python
y
```

```python
smote = SMOTE()
X, y = smote.fit_resample(Samp, y)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X, y)
```

El resultado es un Dataframe con igual número de cada tipo.

### Ejercicio 3

Genera una muestra utilizando el método Reservoir sampling

```python
k = 50
reservoir = []
for i in np.arange(len(liga)):
    if i+1<= k:
        reservoir.append(i)
    else:
        probability = k/(i+1)
        if random.random() < probability:
            reservoir[random.choice(range(0,k))] = i
            
reservoir = pd.DataFrame(reservoir, columns= ['indice']).set_index('indice')

liga_res_sample = reservoir.merge(liga, how='inner', left_index=True, right_index=True)
liga_res_sample   
```
